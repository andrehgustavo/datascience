```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

```
pip install plotly==4.11.0
```

```
data1=pd.read_csv('Season(2019-20).csv')
data2=pd.read_csv('Season(2018-19).csv')
data3=pd.read_csv('nba_extra.csv')
teams = pd.read_csv('teams.csv', usecols=['ABBREVIATION', 'NICKNAME', 'CITY'])
```

```
data1.head()
```

```
season_2019_2020 = data1.iloc[:,[2,7,8]].copy()
season_2019_2020.columns = ["abbreviation", "three_point_made", "three_point_att"]
season_2019_2020["season"] = '2019-2020'
```

```
season_2019_2020.info()
```

```
season_2019_2020.head()
```

```
season_2019_2020.abbreviation.unique()
```

```
#Deixando o nome maiusculo
season_2019_2020.abbreviation = season_2019_2020.abbreviation.str.upper()
season_2019_2020.abbreviation = season_2019_2020.abbreviation.str.replace('NOR', 'NOP')
season_2019_2020.abbreviation = season_2019_2020.abbreviation.str.replace('PHO', 'PHX')
season_2019_2020.abbreviation.unique()
```

```
teams.index = teams.ABBREVIATION
teams.drop(labels="ABBREVIATION", axis=1, inplace=True)
teams
```

```
def getTeamName(abbr):
  return teams.loc[abbr].NICKNAME

def getTeamCity(abbr):
  return teams.loc[abbr].CITY
```

```
season_2019_2020['team_name'] = season_2019_2020.abbreviation.map(getTeamName)
season_2019_2020['city'] = season_2019_2020.abbreviation.map(getTeamCity)
#Reorganizando as colunas
season_2019_2020 = season_2019_2020[['season', 'abbreviation', 'city', 'team_name', 'three_point_made', 'three_point_att']]
season_2019_2020.head()
```

```
season_2019_2020.head()
```

```
stats_2019_2020 = season_2019_2020.groupby(by=['abbreviation', 'city', 'team_name', 'season']).sum()
stats_2019_2020['three_point_perc'] = stats_2019_2020.three_point_made / stats_2019_2020.three_point_att
stats_2019_2020.head()
```

# 2 - Tratando o dataset 2018-2019

```
data2.tail()
```

```
season_2018_2019 = data2.iloc[:,[2,7,8]].copy()
season_2018_2019.columns = ["abbreviation", "three_point_made", "three_point_att"]
season_2018_2019["season"] = '2018-2019'
season_2018_2019.abbreviation = season_2018_2019.abbreviation.str.upper()
season_2018_2019.abbreviation = season_2018_2019.abbreviation.str.replace('NOR', 'NOP')
season_2018_2019.abbreviation = season_2018_2019.abbreviation.str.replace('PHO', 'PHX')
season_2018_2019.info()
```

```
season_2018_2019.head()
```

```
season_2018_2019['team_name'] = season_2018_2019.abbreviation.map(getTeamName)
season_2018_2019['city'] = season_2018_2019.abbreviation.map(getTeamCity)
season_2018_2019 = season_2018_2019[['season', 'abbreviation', 'city', 'team_name', 'three_point_made', 'three_point_att']]
season_2018_2019.head()
```

```
stats_2018_2019 = season_2018_2019.groupby(by=['abbreviation', 'city', 'team_name', 'season']).sum()
stats_2018_2019
```

```
stats_2018_2019['three_point_perc'] = stats_2018_2019.three_point_made / stats_2018_2019.three_point_att
stats_2018_2019
```

# 3 - Tratando dataset3


```
data3.tail(10)
```

```
season_2017_2018 = data3.iloc[:,[4,11,12]].copy()
```

```
season_2017_2018.Tm.unique()
```

TOT - significa total, isto é, o jogador jogou por 2 ou mais times na temporada, e o TOT é a pontuação cumulativa por todos os times. Dessa forma, será excluído do dataset.

```
season_2017_2018 = season_2017_2018[season_2017_2018.Tm != 'TOT']
season_2017_2018.tail(10)
```

```
season_2017_2018.Tm.unique()
```

```
season_2017_2018.columns = ["abbreviation", "three_point_made", "three_point_att"]
season_2017_2018["season"] = '2017-2018'
season_2017_2018["abbreviation"]=season_2017_2018["abbreviation"].str.replace('BRK', 'BKN')
season_2017_2018["abbreviation"]=season_2017_2018["abbreviation"].str.replace('CHO', 'CHA')
season_2017_2018["abbreviation"]=season_2017_2018["abbreviation"].str.replace('PHO', 'PHX')
```

```
season_2017_2018.head()
```

```
season_2017_2018['team_name'] = season_2017_2018.abbreviation.map(getTeamName)
season_2017_2018['city'] = season_2017_2018.abbreviation.map(getTeamCity)
season_2017_2018 = season_2017_2018[['season', 'abbreviation', 'city', 'team_name','three_point_made', 'three_point_att']]
season_2017_2018.head()
```

```
stats_2017_2018 = season_2017_2018.groupby(by=['abbreviation', 'city', 'team_name', 'season']).sum()
stats_2017_2018
```

```
stats_2017_2018['three_point_perc'] = stats_2017_2018.three_point_made / stats_2017_2018.three_point_att
stats_2017_2018
```

# 4 - Concat()


```
df_concat = pd.concat([stats_2019_2020, stats_2018_2019,stats_2017_2018], axis=0)
df_concat
```

```
stats = df_concat.groupby(by=['abbreviation', 'city', 'team_name', 'season']).mean()
stats.head()
```

```
stats.unstack()
stats.head()
```

# 5 - Gráfico

```
fig = go.Figure(
    data=[
        #2017-2018 SEASON
        go.Bar(
            name="17-18 - 3PA",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_att['2017-2018'],
            offsetgroup=0,
            hovertemplate= '<b>%{x}</b>' + '<br>%{text}',
            text = [f'Attempted: <b>{y}</b>' for y in stats.three_point_att['2017-2018']]
        ),
        go.Bar(
            name="17-18 - 3PM",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_made['2017-2018'],
            offsetgroup=0,
            base=stats.three_point_att['2017-2018'],
            customdata = stats.three_point_made['2017-2018'],
            hovertemplate= 
            '<b>%{x}</b>' + '<br>Made: <b>%{customdata}</b>' + '<br>%{text}',            
            text = [f'PER: <b>{i*100:.2f}%</b>' for i in stats.three_point_perc['2017-2018']]       
            
        ),
        #2018-2019 SEASON
       go.Bar(
            name="18-19 - 3PA",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_att['2018-2019'],
            offsetgroup=1,
            hovertemplate= '<b>%{x}</b>' + '<br>%{text}',
            text = [f'Attempted: <b>{y}</b>' for y in stats.three_point_att['2018-2019']]
        ),
        go.Bar(
            name="18-19 - 3PM",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_made['2018-2019'],
            offsetgroup=1,
            base=stats.three_point_att['2018-2019'],
            customdata = stats.three_point_made['2018-2019'],
            hovertemplate= 
            '<b>%{x}</b>' + '<br>Made: <b>%{customdata}</b>' + '<br>%{text}',            
            text = [f'PER: <b>{i*100:.2f}%</b>' for i in stats.three_point_perc['2018-2019']]       
            
        ),
        #2019-2020 SEASON
        go.Bar(
            name="19-20 - 3PA",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_att['2019-2020'],
            offsetgroup=2,
            hovertemplate= '<b>%{x}</b>' + '<br>%{text}',
            text = [f'Attempted: <b>{y}</b>' for y in stats.three_point_att['2019-2020']]
        ),
        go.Bar(
            name="19-20 - 3PM",
            x=stats.index.get_level_values('abbreviation'),
            y=stats.three_point_made['2019-2020'],
            offsetgroup=2,
            base=stats.three_point_att['2019-2020'],
            customdata = stats.three_point_made['2019-2020'],
            hovertemplate= 
            '<b>%{x}</b>' + '<br>Made: <b>%{customdata}</b>' + '<br>%{text}',            
            text = [f'PER: <b>{i*100:.2f}%</b>' for i in stats.three_point_perc['2019-2020']]       
            
        ),
    ],
    layout=go.Layout(
        title="Comparative between 3-point shots in the 17-18, 18-19 and 19-20 NBA seasons",
        yaxis_title="3-point shots numbers",
        #xaxis_tickangle=-45
    )
)
#fig.write_html("file.html")
fig.show()
```
